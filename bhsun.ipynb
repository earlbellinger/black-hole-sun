```python
plt.plot(DF.star_age, DF.L_BH, label='MESA', lw=2)
plt.plot(DF.star_age, DF.M_BH / 10**-5 * 0.34, label='Clayton', ls='--', lw=2)
plt.semilogx()
plt.semilogy()
plt.axhline(1, ls='--', c='k')
plt.axvline(4.579e9, ls='--', c='k')
plt.legend()
plt.xlabel('age [yr]')
plt.ylabel(r'black hole luminosity [L$_\odot$]')
plt.show()
```
